# ð Data Cleaning

## Objective

The goal of this notebook is to clean raw transaction data to make it suitable for analysis and modeling.

## Dataset Overview

The dataset contains transactional records from an online retail store, including product details, customer IDs, quantities, and pricing.

## Key Issues in Raw Data

* Missing CustomerID values
* Negative Quantity (product returns)
* Zero or negative UnitPrice
* Incorrect data types (InvoiceDate)

## Cleaning Steps

1. Remove missing CustomerID
2. Remove negative and zero values
3. Convert InvoiceDate to datetime
4. Create TotalPrice feature

## Output

A cleaned dataset ready for analysis and feature engineering.


**STEP 1 â LOAD DATA**

```python
import pandas as pd

df = pd.read_csv('/content/Online Retail Data Set.csv', encoding='ISO-8859-1')

df.head()
```

**STEP 2 â UNDERSTAND DATA**

```python
df.shape
```

```python
df.info()
```

**STEP 3 â CLEAN DATA**

```python
# Remove missing CustomerID
df = df.dropna(subset=['CustomerID'])

# Remove returns (negative quantity)
df = df[df['Quantity'] > 0]

# Remove invalid price
df = df[df['UnitPrice'] > 0]
```

**STEP 4 â VERIFY CLEANING**

```python
df.shape
df.isnull().sum()
```

```python
(df['Quantity'] <= 0).sum()
(df['UnitPrice'] <= 0).sum()
```

**STEP 5 â CREATE FEATURE**

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```
